##W3A2 Hoán đổi hai số không sd biến tạm thời
Sử dụng phép toán XOR bitwise để hoán đổi giá trị của hai biến. 

```python
a = int(input("Nhập a: "))
b = int(input("Nhập b: "))

a = a ^ b
b = a ^ b
a = a ^ b

print("Sau khi hoán đổi:")
print("a =", a)
print("b =", b)
```
